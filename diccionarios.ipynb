‚òù Un Diccionario es una estructura de datos compuesta que guarda informaci√≥n en pares: <br>


*   Una clave (**key**)
*   Un valor (**value**)



# Conceptos b√°sicos

```python
# Estructura b√°sica
diccionario = {
    "clave1": "valor1",
    "clave2": "valor2",
    "clave3": "valor3"
}
```

‚òπ **Sintaxis:** recuerda que si cambias el orden o los s√≠mbolos, el programa va a dar error. Conceptos clave: <br>


*   Se usan llaves **{ }** para encerrar todo el diccionario
*   Cada elemento se escribe como **clave:valor**
*   Cada par **clave:valor** se separa con una coma (**,**)
*   Las **claves** deben ser √∫nicas (no se pueden repetir)



```python
# se imprime el diccionario
print(diccionario)
print(type(diccionario))
```

```python
# Ejemplo 1
material = {
    "nombre":"cemento",
    "unidad":"kg",
    "cantidad":20,
    "precio":9500,
    "disponible":True
}
```

**Concatenando diccionarios**

¬øQu√© hace el operador |? üëâ Lo que hace es crear un nuevo diccionario con todas las claves y valores de los originales.

```python
dicA = {1:10, 2:20}
dicB = {3:30, 4:40}
dicC = {5:50, 6:60}

# fusiona los tres diccionarios sin modificar los originales.
dictTotal = dicA | dicB | dicC
print(dictTotal)
print(type(dictTotal))
```

```python
# El m√©todo update() tambien combina diccionarios
# Solo que modifica el diccionario original
dicD = {1:10, 2:20}
dicE = {3:30, 4:40}
print("Versi√≥n original dicD:", dicD)
dicD.update(dicE)    # ¬°modifica dicA!
print("Versi√≥n modificada dicD:",dicD)
# Nota que no crea un nuevo diccionario,
# sino que a√±ade los pares del otro al existente.
```

# Copias

```python
# Diccionario original
dict1 = {"dato1": 345, "dato2": 456}
print("Diccionario original:", dict1)

# Creando la copia
dict2 = dict1.copy()
print("Diccionario copiado:", dict2)

#Consultando identidades
print("¬øAmbos diccionarios son iguales?:",dict1 == dict2)
# id devuelve la identidad de un objeto
# se puede comprobar que son objetos diferentes
print("Identidad de dict1:", id(dict1))
print("Identidad de dict2:", id(dict2))
```

```python
# Mira lo que ocurre si altero el orden
dict3 = {"dato1": 345, "dato2": 456}
dict4 = {"dato2": 456, "dato1": 345}
print("¬øAmbos diccionarios son iguales?:",dict3 == dict4)

# Los diccionarios no dependen del orden de las claves cuando se comparan
# Python compara los pares clave-valor, no su posici√≥n
```

```python
# Mira lo que ocurre si no se hace una copia
dict5 = {"dato1": 345, "dato2": 456}
dict6 = dict5
print("¬øAmbos diccionarios son iguales?:",dict5 == dict6)
print("Identidad de dict5:", id(dict5))
print("Identidad de dict6:", id(dict6))
```

```python
# Cuando definimos dict5 = dict6 no se crea una copia nueva
# En realidad ambas variables apuntan al mismo diccionario en memoria
# Mira la consecuencia de esta acci√≥n:

print("dict5 original:", dict5)
print("dict6 original:", dict6)

#Solo modifico un diccionario
dict5["dato1"] = 123
# Mira lo que ocurre con el otro, tambi√©n se modifica
print(dict5)
print(dict6)
# Esto ocurre porque en realidad ambos son el mismo objeto
# Si realmente quieres hacer una copia independiente usa .copy()
```

# Recorrido

```python
# Podemos consultar por una clave en espec√≠fico
if "precio_unitario" in material:
  print(material["precio_unitario"])
else:
  print("No definido")
```

```python
if "proveedor" not in material:
  print("La clave 'proveedor' no est√° registrada todav√≠a.")
```

```python
# Ahora consulto solo claves
for clave in material:
  print(clave)
```

```python
# Solo valores
for valor in material.values():
  print(valor)
```

```python
# Claves y valores juntos
for clave, valor in material.items():
    print(clave, ":", valor)
```

‚ö° Nota que en los recorridos de las claves con **for**, usamos solo el nombre del diccionario (por ejemplo, for clave in material:) y no algo como "material.clave". En ese caso, por defecto, Python entiende que queremos recorrer las claves del diccionario.

‚òπ **while** no recorre directamente los diccionarios, primero debe convertirse sus elementos a listas. M√°s adelante te muestro un ejemplo de c√≥mo hacerlo. De momento, sigamos comprendiendo las operaciones b√°sicas que podemos hacer con los diccionarios.

‚Ü™ Ejemplo recorrido

```python
# Se crea el diccionario
tabla_pagos = { "actividad-1": 23,
                "actividad-2": 45,
                "actividad-3": 33,
                "actividad-4": 98,
                "actividad-5": 56,
                "actividad-6": 66,
                "actividad-7": 99
                }
print(tabla_pagos)
```

# Ejemplo de aplicaci√≥n

‚òù Mira que interesante lo que podemos hacer si integramos algunos de los conceptos que hemos aprendido. Vamos a suponer que tenemos un pedido con cantidad de cada material. Podemos calcular el costo total del proyecto de la siguiente manera:

```python
pedido = [
    {"nombre": "Cemento", "cantidad": 40, "precio_unitario": 9500},
    {"nombre": "Arena", "cantidad": 5, "precio_unitario": 15500},
    {"nombre": "Grava", "cantidad": 7, "precio_unitario": 17000}
]

#Defino costo inicial
total = 0
for item in pedido:
    # se calcula el subtotal de un material
    subtotal = item["cantidad"] * item["precio_unitario"]
    # forma abreviada de: total = total + subtotal
    # se va acumulando el monto total
    total += subtotal
    print(f"{item['nombre']}: {subtotal} colones")

print(f"\nCosto total del pedido: {total} colones")
```
