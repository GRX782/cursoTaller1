✌A continuación te voy a explicar qué es una **lista** en Python, cómo se crean y se modifican y para qué son útiles.

**Sintaxis básica:** nombre_lista = [elemento1, elemento2, elemento3]

```python
# Las listas se crean escribiendo los elementos entre corchetes [ ], separados por comas
numeros = [20,30,48.6,26.9 ]
nombres = ["Eduardo","Felipe","María"]
mixta = [True, 25,"hola",35.6]

```

```python
# La lista es una colección que permite almacenar varios elementos dentro de una misma variable
print(numeros)
print(type(numeros))
print(len(numeros))
```

```python
print(nombres)
print(type(nombres))
print(len(nombres))
```

```python
print(mixta)
print(type(mixta))
print(len(mixta))
```

⚡También se pueden crear **listas vacías** y llenarlas después.

Las listas se consideran "**mutables**" dado que se pueden modificar.

```python
datos = [ ] #lista vacía
# se agregan elementos con la función "append()"
print("Versión original:",datos)
datos.append(43)
datos.append(64)
datos.append(156)
print("Versión 2:",datos)

# Nota que append() agrega un elemento al final de la lista existente
datos.append(5)
datos.append(28)
print("Versión 3:",datos)
```

✍ Similar a lo que aprendimos con las cadenas de caracteres, en las listas cada elemento tiene una posición (índice) que comienza en cero.

```python
# se indica entre corchetes el índice que se desea consultar
print(nombres[0])
print(nombres[1])
print(nombres[2])
```

```python
# también se puede consultar de derecha a izquierda
print(numeros)
print(numeros[-1])
print(numeros[-2])
print(numeros[-3])
print(numeros[-4])
```

✅ Ahora vamos a modificar elementos en específico

```python
print(numeros)
numeros[0] = 40
print(numeros)
```

```python
print(numeros)
numeros[-2] = 100
print(numeros)
```

Anteriormente, vimos que **append()** agrega un elemento al final de la lista.

Por su parte, **insert()** agrega un elemento en una posición en específico:

```python
# arena posee el índice 0
# piedra posee el índice 1
# cemento posee el índice 2 o -1 (desde el final)
materiales = ["arena","piedra","cemento"]
print(materiales)
# agregamos elementos a la lista
materiales.append("hierro") # lo coloca al final de la lista
materiales.insert(1,"madera") # lo va colocar justo después de arena
# Verificamos la versión final de la lista
print(materiales)
```

⛹ Ahora vamos a **eliminar elementos** de la lista:

```python
# te presento la función remove():
print(materiales)
# agrego el mismo elemento para que se duplique
materiales.append("madera")
print(materiales)
# nota como remove() solo elimina la primera coincidencia
materiales.remove("madera")
print(materiales)
```

```python
# también puedes utilizar la función pop()
print(materiales)
# se le indica el indice que desea eliminar
materiales.pop(2) # en este caso va a eliminar "cemento"
print(materiales)
```

⚡**pop()** aunado a eliminar el elemento de la lista lo devuelve (retorna) para que puedas usuarlo o guardarlo en una variable

```python
numeros = [10, 20, 30, 40]
valor = numeros.pop(2)   # elimina el elemento en el índice 2
print(valor)             # imprime 30
print(numeros)           # la lista ahora es [10, 20, 40]
```

☹ Por su parte, la sentencia **del** también elimina por índice pero NO retorna el valor. Nota que la sintaxis es muy distinta respecto a **pop()**

```python
numeros = [10, 20, 30, 40]
del numeros[2]            # elimina el elemento en la posición 2
print(numeros)            # la lista ahora es [10, 20, 40]
```

```python
# Podemos borrar todos los datos a partir de un índice en específico
numeros = [10, 20, 30, 20, 40, 20, 50]
print("Lista original:", numeros)
del numeros[3:]
print("Después de eliminar desde el índice 3:", numeros)
```

```python
# Supongamos ahora que tienes una lista con valores repetidos
# Quieres eliminar todos los elementos iguales a un valor específico a la vez
numeros = [10, 20, 30, 20, 40, 20, 50]
print("Lista original:", numeros)

# Defino el valor que deseo eliminar
valor = 20

# Bucle para repetir remove()
while valor in numeros:
    numeros.remove(valor)

print("Después de eliminar todos los 20:", numeros)
```

**Operaciones comunes**

☝ Similar a cadena de caracteres, con las listas también podemos ejecutar algunas operaciones útiles, mira:

```python
# podemos contar cuántos elementos hay
print(materiales)
print(f"La lista tiene {len(materiales)} elementos")
```

⏰ **Recuerda:** no debes confundir la **cantidad de elementos** len() con los **índices** que comienzan en cero y se utilizan oara acceder a un elemento mediante su **posición**

```python
# Mostrar la lista completa
print("Lista completa:", materiales)

# Mostrar los índices disponibles
print("\nÍndices válidos: 0, 1 , 2 y 3")
print("materiales[0] =", materiales[0])
print("materiales[1] =", materiales[1])
print("materiales[2] =", materiales[2])
print("materiales[3] =", materiales[3])
```

```python
# mira lo que ocurre si intentas acceder a una posición que no existe
print("\nSi intentamos acceder a materiales[4]...")
try:
    print(materiales[4])
except IndexError:
    print("⚠️ Error: el índice 4 no existe, porque los índices van de 0 a 3.")
```

```python
# Podemos contar cuántas veces se repite un valor en una lista
numeros = [10, 20, 30, 20, 40, 20, 50]
print("Lista original:", numeros)
cantidad = numeros.count(20)
print("Cantidad de veces que se repite el número 20 en la lista:", cantidad)
```

☝ Continuando con las operaciones:

**Pertenencia**

```python
# puedes verificar si un elemento en particular existe
# para ello se utilizan los operadores "in" y "not in"
# devuelven True o False
print(materiales)
# puedo personalizar la consulta
print("Existe \"grava\" en materiales:","grava" in materiales) # False
# o sencillamente verificarlo
print("hierro" in materiales) # True
print("arena" not in materiales) # False
```

```python
# Podemos compinar estos operadores con herramientas de control de flujo:

# Lista de materiales disponibles
materiales = ["cemento", "arena", "grava", "acero"]

# Solicitar al usuario un material
material = input("Ingrese el material que desea verificar: ").lower()

# Verificar pertenencia
if material in materiales:
    print(f"✅ Sí, {material} está disponible en la lista.")
elif material not in materiales:
    print(f"❌ No, {material} no está en la lista de materiales.")
else:
    print("⚠️ Verifique el dato ingresado.")
```

```python
# ahora vamos a contar cuantas veces se repite un valor
datos = [10, 20, 30, 40, 30, 50, 30]
print(f"El numero 30 se repite {datos.count(30)} veces en la lista")
```

**Recorrido**

```python
# acá te muestro como podemos recorrer las listas
frutas = ["manzana", "pera", "uva"]

for i in frutas:
    print(i)
```

```python
# Otro ejemplo
# Lista de materiales de construcción
materiales = ["cemento", "arena", "grava", "acero"]

print("=== Recorriendo la lista con FOR ===")
for i in materiales:
    print(f"• Material: {i}")

print("\n=== Recorriendo la lista con WHILE ===")
i = 0  # contador o índice
while i < len(materiales):
    print(f"• Material: {materiales[i]}")
    i += 1  # importante: incrementar el contador
```

**Concatenación y repetición**

```python
# Se pueden unir listas o repetir sus elementos
a = [1, 2]
b = [3, 4]
print(a + b)    # [1, 2, 3, 4]  → concatenación
print(a * 3)    # [1, 2, 1, 2, 1, 2]  → repetición
```

```python
# Puedo crear listas nuevas

# Listas base
a = [1, 2]
b = [3, 4]

print("Lista a:", a)
print("Lista b:", b)

# Concatenar listas: se unen una detrás de otra
c = a + b
print("Lista c (concatenada):", c)

# Repetir una lista varias veces
d = a * 3
print("Lista d (repitiendo a tres veces):", d)
```

**Sublistas [inicio:fin]**

```python
# se puede extrar una sublista de una lista

# Se define la lista
lista_numeros = [10, 20, 30, 40, 50]
print(lista_numeros)
# Recuerda que los índices de la lista son:
#   0 → 10
#   1 → 20
#   2 → 30
#   3 → 40
#   4 → 50

# ten presente que incluye el índice inicial
# pero excluye el índice final
# "desde ...., hasta uno antes de ..."
sublista1 =  lista_numeros[1:3]
print(sublista1)
sublista2 =  lista_numeros[3:4]
print(sublista2)

# si dejas vacío el inicio Python asume que inicio en cero
sublista3 =  lista_numeros[:2]
print(sublista3)

# si dejas vacío el fin Python toma hasta el final de la lista
# no excluye nada
sublista4 =  lista_numeros[2:]
print(sublista4)
```

**Listas anidadas**

```python
# Lista anidada (lista de listas)

# Partimos sobre el hecho que tenemos varias listas
lista1 = ["cemento", 25, "sacos"]
lista2 = ["arena", 10, "m³"]
lista3 = ["grava", 8, "m³"]

# Ahora creamos una lista de listas
control_inventario = [lista1, lista2, lista3]

# La lista "control_inventarios" contiene tres listas anidadas
print("Lista de listas:", control_inventario)
print(f"Posee {len(control_inventario)} listas anidadas")
```

```python
# Acceso a elementos
print("Esta es la sublista 1:", control_inventario[0])      # Consulta la lista anidada según el índice

# Tambien se pueden hacer consultas mas específicas
# El primer índice [ ] selecciona cuál sublista quiero consultar.
# El segundo índice [ ] selecciona el elemento dentro de esa sublista
print(f"El valor ubicado en la sublista 1, específicamente en el índice 2 es: {control_inventario[0][2]}")  # sacos
```

☣ Si bien las **listas anidadas** son fáciles de crear, si cambia el orden de las sublistas el código se rompe y son difíciles de mantener si hay muchos datos o campos, en especial porque la sintaxis no dice mucho.
