Una **clase** es como un molde, plantilla o modelo que describe cómo serán los objetos creados a partir de esa clase defiendo qué datos tendrán (**atributos**) junto con qué acciones pueden realizar (**métodos**)

Sintáxis de definición de clases

```python
# Estructura básica

class NombreClase:
  pass   #Clase vacía por ahora
```

⚡ Un **"método"** es una función que está definida **dentro de una clase** y sirve para realizar acciones relacionadas con los objetos creados a partir de esa clase

```python
class Usuario:
    def saludar(self):
        print("Hola, soy un usuario")
```

```python
# creo el objeto
pedrito =  Usuario()
```

```python

# Invoco al método "saludar()" que vive dentro de la clase "Usuario"
pedrito.saludar()
```

**self** es el nombre que se le da al objeto que está ejecutando el método. Te permite acceder a los atributos del objeto

**\_\_init\_\_** es un método especial de Python que se ejecuta automáticamente cada vez que se crea un objeto.

☝Cuando creas un objeto, necesitas "llenarlo" con datos iniciales. **\_\_init\_\_** es el encargado de inicializar esos datos. Por eso su nombre: init → initialize → inicializar

❗La mejor forma de comprender esto es con ejemplos:

**Ejemplo 01**

```python
# Se crea la clase
class Usuario:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años")
```

```python
# Se crean objetos a partir de la clase creada
usuario1 = Usuario("Juan", 25)
usuario2 = Usuario("María", 30)
usuario3 = Usuario("Pedro", 35)
```

```python
# Consulto los atributos
# No uso paréntesis porque no son métodos
print(usuario1.nombre, usuario1.edad)
print(usuario2.nombre, usuario2.edad)
print(usuario3.nombre, usuario3.edad)
```

```python
# Consulto las acciones (métodos)
usuario1.saludar()
usuario2.saludar()
usuario3.saludar()
```

**Ejemplo 2**

```python
class datos:
  # atributos
  def __init__(self, nombre, edad, correo,telefono):
    self.nombre = nombre
    self.edad = edad
    self.correo = correo
    self.telefono = telefono

  # métodos
  def saludar(self):
    print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años")
  def print_info(self):
    print(f"Nombre: {self.nombre}")
    print(f"Edad: {self.edad}")
    print(f"Correo: {self.correo}")
    print(f"Teléfono: {self.telefono}")
```

```python
# Creo los objetos
cliente1 = datos("Juan", 25, "jsolano@correo.com", "8888-8888")
cliente2 = datos("María", 30, "mgarcia@correo.com", "9999-9999")
cliente3 = datos("Pedro", 35, "pdiaz@correo.com", "7777-7777")
cliente4 = datos("Ana", 28, "asanchez@correo.com", "6666-6666")
```

```python
# invoco los métodos
cliente1.saludar()
cliente2.saludar()
cliente3.saludar()
cliente4.saludar()
```

```python
# Consulto los datos
print("Base de datos clientes 2025")
print("-"*10)
cliente1.print_info()
print("-"*10)
cliente2.print_info()
print("-"*10)
cliente3.print_info()
print("-"*10)
cliente4.print_info()
```

**Ejemplo 3**

```python
class carros:
  # atributos
  def __init__(self, litros,velocidad):
    self.litros = litros
    self.velocidad = velocidad

  # métodos
  def arrancar(self):
    if self.litros > 0:
      print("Puede arrancar el carro")
    else:
      print("No hay suficiente combustible")
  def acelerar(self):
    print(f" La velocidad inicial es {self.velocidad}")
    incremento = float(input(" Ingrese el incremento de velocidad: "))
    self.velocidad += incremento
    print(f" La velocidad final es {self.velocidad} km/h")
```

```python
# Se crean objetos
carro1 = carros(10,40)
carro2 = carros(0,60)
carro3 = carros(5,50)
carro4 = carros(15,20)
```

```python
# invoco los métodos
carro1.arrancar()
carro2.arrancar()
carro3.arrancar()
carro4.arrancar()
```

```python
# Nota que este método no requiere que le ingrese argumentos
carro1.acelerar()
```

**Ejemplo 4**

```python
# definimos la clase
class material:
  # atributos
  def __init__(self, nombre, precio):
    self.nombre = nombre
    self.precio = precio
```

```python
# creamos varios objetos a partir de la clase creada
m1 = material("cemento", 8500)
m2 = material("arena", 3000)
m3 = material("grava", 5000)
```

```python
# almacenamos los objetos dentro de una tupla (inmutable)
materiales = (m1,m2,m3)
```

```python
# recorremos la tupla con un bucle
print("Lista de Materiales")
for i in materiales:
  print(f"  {i.nombre} - ₡{i.precio}")
```

```python
# podemos hacer operaciones matemáticas
total = 0

for j in materiales:
  total += j.precio
print(f"El costo total de los materiales es: ₡{total}")
```
